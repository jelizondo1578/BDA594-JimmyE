Write a function that can compute the difference between two temperatures.  Assume the

Assume the first argument is obligatory and then the second argument is optional.  When the second argument is omitted, just return the result of converting the first argument to Celsius.
Show that you have checked your work, using all of the following examples.

Here is how the function should work:

```
    In [ ]:  fahrenheit_to_celsius_diff(212,32)
    Out [ ]: 100
    In [ ]: fahrenheit_to_celsius_diff(-40, 212)
    Out [ ]: -140.0
    In [ ]:  fahrenheit_to_celsius_diff(212)
    Out [ ]: 100
    In [ ]:  fahrenheit_to_celsius_diff(212, 0)
    Out [ ]: 117.78

```

### Solution

````python

def fahrenheit_to_celsius_diff(temp1, temp2=None):
    if temp2 is None:
        # If only one argument is given, just convert it
        return fahrenheit_to_celsius(temp1)
    else:
        # Compute the difference in Celsius
        c1 = fahrenheit_to_celsius(temp1)
        c2 = fahrenheit_to_celsius(temp2)
        return c1 - c2
````

````python
def test(values):
    """
    Check if a Sudoku grid is legal.
    Print 'Valid!' if legal, 'Invalid!' if illegal.
    """
    for s in squares:
        val = values[s]
        if val in digits:  # <- colon is required here
            # If any peer has the same value, the grid is invalid
            for peer in peers[s]:
                if values[peer] == val:
                    print("Invalid!")
                    return
    print("Valid!")
````
